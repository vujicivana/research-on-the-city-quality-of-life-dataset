```python
from scipy.stats import pearsonr
```

```python
lista_nezavisnih=['Housing']
for kolona1 in ['Cost of Living', 'Startups', 'Venture Capital', 'Travel Connectivity', 'Commute', 'Business Freedom', 'Safety', 'Healthcare', 'Education', 'Environmental Quality', 'Economy', 'Taxation', 'Internet Access', 'Leisure & Culture', 'Tolerance', 'Outdoors']:
    d1=data[kolona1]
    zavisni=False
    for kolona2 in lista_nezavisnih:
        if kolona1==kolona2:
            continue
        d2=data[kolona2]
        corr, _=pearsonr(d1, d2)
        if corr<-0.1 or corr>0.1:
            print("col1 %s i col2 %s corr %.3f" % (kolona1, kolona2, corr))
            zavisni=True
            break
    if not zavisni:
        lista_nezavisnih.append(kolona1)

print(lista_nezavisnih)

data2=data[lista_nezavisnih]
```

# Klasterizacija

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
def get_city(index):
    return data.loc[index]['UA_Name']

def get_clusters(l):
    clusters={}
    for i, c in enumerate(l):
        if not c in clusters:
            clusters[c]=[]
        city=get_city(i)
        clusters[c].append(city)
    return clusters
```

# KMeans

```python
def kmeans_clusters(listData):
  d={}
  for k in range(2, 5):
    kmeans=KMeans(n_clusters=k).fit(listData)
    d[k]=kmeans
  return d

def call_kmeans(listData):
  kmeans=kmeans_clusters(listData)
  for k, km in kmeans.items():
    print()
    print("K=", k)
    listClusters=km.labels_
    data_for_DataFrame['Klaster'].append('KMeans')
    data_for_DataFrame['Broj klastera'].append(k)
    ss=silhouette_score(listData, listClusters)
    data_for_DataFrame['Silhouette Score'].append(round(ss, 2))
    km=get_clusters(listClusters)
    for cluster, list_ in km.items():
      print(cluster, ":", list_)
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
def agglomerative_clusters(ListData):
  d={}
  for k in range(2, 5):
    agglomeratives=AgglomerativeClustering(n_clusters=k, linkage='complete').fit(ListData)
    d[k]=agglomeratives
  return d

def call_agglomerative(listData):
  agglomerative=agglomerative_clusters(listData)
  for k, km in agglomerative.items():
    print()
    print("K=", k)
    listClusters=km.labels_
    data_for_DataFrame['Klaster'].append('Agglomerative')
    data_for_DataFrame['Broj klastera'].append(k)
    ss=silhouette_score(listData, listClusters)
    data_for_DataFrame['Silhouette Score'].append(round(ss, 2))
    km=get_clusters(listClusters)
    for cluster, list_ in km.items():
      print(cluster, ":", list_)
```

```python
data_for_DataFrame={'Klaster':[], 'Broj klastera':[], 'Silhouette Score':[]}
```

```python
call_agglomerative(data1)
```

```python
pd.DataFrame(data=data_for_DataFrame)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
def dbscan_clusters(listData):
  d={}
  for k in (4, 6, 8):
    dbscan=DBSCAN(eps=10, min_samples=k).fit(listData)
    d[k]=dbscan
  return d

def call_dbscan(listData):
  dbscan=dbscan_clusters(listData)
  for k, km in dbscan.items():
    print()
    print("K=", k)
    listClusters=km.labels_
    data_for_DataFrame['Klaster'].append('DBSCAN')
    data_for_DataFrame['Min Samples'].append(k)
    ss=silhouette_score(listData, listClusters)
    data_for_DataFrame['Silhouette Score'].append(round(ss, 2))
    km=get_clusters(listClusters)
    for cluster, list_ in km.items():
      print(cluster, ":", list_)
```

```python
data_for_DataFrame={'Klaster':[], 'Min Samples':[], 'Silhouette Score':[]}
```

```python
call_dbscan(data1)
```

```python
pd.DataFrame(data=data_for_DataFrame)
```

# Mean Shift

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
eb=estimate_bandwidth(data1)
eb
```

```python
def ms_clusters(ListData):
  d={}
  for k in range(8, 11):
    ms=MeanShift(bandwidth=k).fit(ListData)
    d[k]=ms
  return d

def call_ms(listData):
  ms=ms_clusters(listData)
  for k, km in ms.items():
    print()
    print("K=", k)
    listClusters=km.labels_
    data_for_DataFrame['Klaster'].append('MeanShift')
    data_for_DataFrame['Bandwidth'].append(k)
    ss=silhouette_score(listData, listClusters)
    data_for_DataFrame['Silhouette Score'].append(round(ss, 2))
    km=get_clusters(listClusters)
    for cluster, list_ in km.items():
      print(cluster, ":", list_)
```

# Spectral Clustering

# Gaussian Mixture

```python
from sklearn.mixture import GaussianMixture
```

```python
def gmm_clusters(ListData):
  d={}
  for k in range(2, 5):
    gmm=GaussianMixture(n_components=k).fit(ListData)
    d[k]=gmm
  return d

def call_gmm(listData):
  gmm=gmm_clusters(listData)
  for k, km in gmm.items():
    print()
    print("K=", k)
    listClusters=km.predict(listData)
    data_for_DataFrame['Klaster'].append('GMM')
    data_for_DataFrame['Broj klastera'].append(k)
    ss=silhouette_score(listData, listClusters)
    data_for_DataFrame['Silhouette Score'].append(round(ss, 2))
    km=get_clusters(listClusters)
    for cluster, list_ in km.items():
      print(cluster, ":", list_)
```

```python
data_for_DataFrame={'Klaster':[], 'Broj klastera':[], 'Silhouette Score':[]}
```

```python
call_gmm(data1)
```

```python
pd.DataFrame(data=data_for_DataFrame)
```

# Grafik

```python
import numpy as np
import matplotlib.pyplot as plt
rng=np.random.RandomState(60)
transformation=rng.normal(size=(2, 2))
X=np.dot(data2, transformation)
```

# KMeans

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
y_pred=kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='green', s=60, linewidth=2);
ss=silhouette_score(X, y_pred)
print('Silhouette Score is', ss)
dbs=davies_bouldin_score(X, y_pred)
print('Davies Bouldin Score is', dbs)
```

# Agglomerative Clustering

```python
agglomerative=AgglomerativeClustering(n_clusters=3, linkage='complete')
y_pred=agglomerative.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis');
ss=silhouette_score(X, y_pred)
print('Silhouette Score is', ss)
dbs=davies_bouldin_score(X, y_pred)
print('Davies Bouldin Score is', dbs)
```

# DBSCAN

```python
dbscan=DBSCAN(eps=0.9, min_samples=9)
y_pred=dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis');
ss=silhouette_score(X, y_pred)
print('Silhouette Score is', ss)
dbs=davies_bouldin_score(X, y_pred)
print('Davies Bouldin Score is', dbs)
```

# Mean Shift

```python
ms=MeanShift(bandwidth=2)
y_pred=ms.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis');
plt.scatter(ms.cluster_centers_[:, 0], ms.cluster_centers_[:, 1], c='green', s=60, linewidth=2);
ss=silhouette_score(X, y_pred)
print('Silhouette Score is', ss)
dbs=davies_bouldin_score(X, y_pred)
print('Davies Bouldin Score is', dbs)
```

# Spectral Clustering

```python
spectral=SpectralClustering(n_clusters=3)
y_pred=spectral.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis');
ss=silhouette_score(X, y_pred)
print('Silhouette Score is', ss)
dbs=davies_bouldin_score(X, y_pred)
print('Davies Bouldin Score is', dbs)
```

# Gaussian Mixture

```python
gmm=GaussianMixture(n_components=3)
y_pred=gmm.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis');
ss=silhouette_score(X, y_pred)
print('Silhouette Score is', ss)
dbs=davies_bouldin_score(X, y_pred)
print('Davies Bouldin Score is', dbs)
```
